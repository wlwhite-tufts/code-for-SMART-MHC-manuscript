```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
metrics = pd.read_csv(
    '<path/to/your>/selected_designs_with_metrics.csv', #this is generated by ../05_filtering/final_filtering.ipynb
    index_col=0)
print(len(metrics))

seqs = pd.read_csv('full_seqs_dna_300mer.csv',index_col=0)
print(len(seqs))
```

```python
metrics = metrics.sort_values(by='interface_lddt',ascending=False)
metrics.index = range(len(metrics))
```

```python
with open('<desired/path/to>/full_seqs_dna_300mer_sorted.list','w') as f:
    for name in tqdm(metrics['short_name']):
        matches = seqs[seqs['design_name'].map(lambda x: name in x)].sort_values(by='design_name')
        
        if len(matches) != 4 and len(matches) != 8:
            print(name)
            
        if len(matches) == 8:
            if np.random.rand(1) < 0.5:
                matches = matches.iloc[:4,:]
            
        for seq in matches['full_seq']:
            f.write(seq+'\n')
```
