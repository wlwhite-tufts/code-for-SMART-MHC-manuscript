```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

```python
data = pd.read_csv('frag WW.CSV',skiprows=1)
```

```python
data.head()
```

```python
data['Y'] = data['Y(Counts)']/1e6
```

```python
plt.figure(figsize=[16,5])

plt.plot(data['X(Daltons)'],data['Y'],linewidth=1)

masses = [7457, 10198, 10823, 21646, 30478]
for m in masses:
    row = data[data['X(Daltons)']==m]
    l = len(str(m))

    if m == 10823:
        plt.text(x=row['X(Daltons)']+1e2,y=row['Y']-0.1,s=str(m),fontdict={'fontsize':16})
    else:
        plt.text(x=row['X(Daltons)']-(l/4)*6.5e2,y=row['Y']+0.15,s=str(m),fontdict={'fontsize':16})

plt.xlim([5000,35000])
plt.ylim([-0.05,5.8])
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.savefig('MS_fragments.png',dpi=300,bbox_inches='tight')
```
