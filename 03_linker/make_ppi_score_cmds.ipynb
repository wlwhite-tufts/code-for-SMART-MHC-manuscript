```python
with open('<desired/path/to>/ppi_score.cmds','w') as f:
    for i,row in inputs.iterrows():
        name = row['description']
        binder_range = f'1-{row["linker_start"]-1}'
        MHC_range = f'{row["linker_end"]+2}-{len(row["sequence"])}'
        linker_range = f'{row["linker_start"]}-{row["linker_end"]+1}'
        f.write('<path/to/your>/rosetta/latest/bin/rosetta_scripts @score_only.flags '+\ #your rosetta installation
                '-parser:protocol all_master_scores_fused.xml '+\
                f'-in:file:s <path/to/your>/ppi_score_inputs/{name}.pdb -out:file:score_only <desired/path/to>ppi_scores/{name}.sc '+\
                f'-parser:script_vars binder_range={binder_range} MHC_range={MHC_range} linker_range={linker_range}\n')
```
